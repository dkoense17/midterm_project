# Which countries are the most responsible for CO2 emissions?

Below is an analysis breaking down which countires are the most responsible:

```python
co2 = pd.read_csv("owid-co2-data.csv")
co2 = co2[co2["year"] >= 2000] 
co2 = co2[co2["iso_code"] != "OWID_WRL"]
averages = co2.groupby('iso_code').mean() #using iso_code in this dataset because continent data does not have an iso_code
averages = averages.sort_values(by=['co2'], ascending = False)
averages.head(10)
```

```python
co2 = pd.read_csv("owid-co2-data.csv")
co2 = co2[co2["year"] >= 2000] 
co2 = co2[co2["iso_code"] != "OWID_WRL"]
co2.groupby("iso_code")["co2"].mean().sort_values(ascending=False).head(25)
#tried it this way so that I did not see every single data point, just the CO2
```

Using this chart we can see that there are some countries that average far more use in the 21st century. The numerical data point here represents the average ouput of CO2 in the 21st century in million tons. Choosing only those countries that averaged over one thousand, this line plot was created that shows CO2 consumption per year since the year 2000:

```python
co2 = pd.read_csv("owid-co2-data.csv")
co2 = co2[co2["year"] >= 2000]
co2.set_index("country", inplace = True)

co2China = co2.loc["China"]
x1 = co2China["year"]
y1 = co2China["co2"]

co2USA = co2.loc["United States"]
x2 = co2USA["year"]
y2 = co2USA["co2"]

co2India = co2.loc["India"]
x3 = co2India["year"]
y3 = co2India["co2"]

co2Russia = co2.loc["Russia"]
x4 = co2Russia["year"]
y4 = co2Russia["co2"]

co2Japan = co2.loc["Japan"]
x5 = co2Japan["year"]
y5 = co2Japan["co2"]

plt.figure(figsize=(9,6))
plt.plot(x1,y1,label="China")
plt.plot(x2,y2,label="United States")
plt.plot(x3,y3,label="India")
plt.plot(x4,y4,label="Russia")
plt.plot(x5,y5,label="Japan")
plt.legend()
plt.show()
```

This next line plot shows the yearly percent change in those 5 countries since the year 2000:

```python
y1 = co2China["co2_growth_prct"]
y2 = co2USA["co2_growth_prct"]
y3 = co2India["co2_growth_prct"]
y4 = co2Russia["co2_growth_prct"]
y5 = co2Japan["co2_growth_prct"]

plt.figure(figsize=(9,6))
plt.plot(x1,y1,label="China")
plt.plot(x2,y2,label="United States")
plt.plot(x3,y3,label="India")
plt.plot(x4,y4,label="Russia")
plt.plot(x5,y5,label="Japan")
plt.legend()
plt.show()
```
